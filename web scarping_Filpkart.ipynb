```python
table=[['Name','Rating','Price']]
for laptop in laptops:
    name=laptop.find_all('div',{'class':'_3wU53n'})
    name=name[0].text
    name=name.split(' - ')[0]
  # if is rating is null then if statement solve that issue  
    rating=laptop.find_all('div',{'class':'hGSR34'})
    if rating:
        rating=rating[0].text
    else:
        rating='-'
    
    price=laptop.find_all('div',{'class':'_1vC4OE _2rQ-NK'})
    price=price[0].text.replace('₹','').strip().replace(',','')
    
    table.append([name,rating,price])
    
    
    
```
